```python
def convert3(obj):
    L=[]
    counter=0
    for  i in ast.literal_eval(obj):
        if counter!=3:
            L.append(i["name"])
            counter+=1
        else:
            break
    return L
```

From this we can only need the director

```python
def fetch_director(obj):
    L=[]
    for  i in ast.literal_eval(obj):
        if i["job"]=="Director":
            L.append(i["name"])
        else:
            break
    return L
```

```python
movies['crew']=movies['crew'].apply(fetch_director)
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    #now convertinng this list into String
    return " ".join(y)
```

```python
new_df["tags"]=new_df["tags"].apply(stem)
```

```python
# Now we are again doing CountVectorization
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000, stop_words="english")
```

```python
vectors= cv.fit_transform(new_df["tags"]).toarray()
```

```python
cv.vocabulary_
```

Now Our all the movies have been converted into Vectors Now we can calculate the cosine Distance here

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity= cosine_similarity(vectors)
```

```python
cosine_similarity(vectors).shape
```

The Workflow  is we will firstly  find out the Index of the Given movie. Through this index 
we will move into the similarity matrix and then sort these and then fetch the Top 5 movies

```python
def recommend(movie):
    movie_index= new_df[new_df['title']==movie].index[0]
    distances= similarity[movie_index]
    movies_list=sorted(list(enumerate(distances)), reverse=True, key=lambda x:x[1])[1:6]
    for i in movies_list:
        print(new_df.iloc[i[0]].title)

```

```python
recommend("Batman Begins")
```
